# Heart Desease classification project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df
```

```python
x, y = df.drop('stroke', axis=1), df['stroke']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

def evaluate_model(true, predicted):
    acc = accuracy_score(true, predicted)
    f1 = f1_score(true, predicted)
    precision = precision_score(true, predicted)
    recall = recall_score(true, predicted)
    return acc, f1, precision, recall
```

```python
models = {
    "K-Neighbors Classifire": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest Classifire": RandomForestClassifier(),
    "Support Vector Machine": SVC()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train, y_train) # Train model

    # Make predictions
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    # Evaluate Train and Test dataset
    acc, f1, precision, recall = evaluate_model(y_test, y_test_pred)
    print(list(models.keys())[i])

    print(f"Accuracy: {acc:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")


    print('='*35)
    print('\n')
```

```python
rfparams = {
    'n_estimators': [50, 100, 200],                   # Enough variety, avoid too small (20 often underfits)
    'criterion': ['gini', 'entropy'],                 # 'log_loss' only works if `probability=True` or in newer versions
    'max_features': ['sqrt', 'log2'],                 # 'sqrt' is generally best for classification
    'max_depth': [10, 20, 30, None],                  # Avoid going too deep (overfitting)
    'min_samples_split': [2, 5],                      # 2 is default; 5 helps regularization
    'min_samples_leaf': [1, 2, 4]                     # 1 = very flexible; 4 = better generalization
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
model= RandomizedSearchCV(RandomForestClassifier(class_weight='balanced'), rfparams, cv=5,scoring='recall',n_jobs=-1)
model.fit(x_train, y_train)
```

```python
model.best_score_
```

```python
y_test_pred = model.predict(x_test)
y_test_pob = model.predict_proba(x_test)
print(evaluate_model(y_test,y_test_pred))
print(y_test_pob)
```

```python
import pickle
with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```
